```python
import pandas as pd

df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_127006.csv", skiprows=4)
df.head()
```

```python
print(df.columns)
```

```python
data_2020 = df[['Country Name', '2020']].dropna()
data_2020 = data_2020.sort_values(by='2020', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x='2020', y='Country Name', hue='Country Name',data=data_2020, palette='dark', dodge=False, legend=False)
plt.title("Top 10 Countries by Population in 2020")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['2020'].dropna(), bins=20, kde=True, color='dodgerblue')
plt.title("Histogram of Country Populations in 2020")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.bar(data_2020, x='Country Name', y='2020', title='Top 10 Countries by Population (2020)', labels={'2020':'Population'},color_discrete_sequence=['slateblue'])
fig.show()
```
